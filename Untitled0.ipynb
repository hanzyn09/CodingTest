# 1. 그리드 알고리즘
  - 그리드는 탐욕 알고리즘

```python
n = 1260
count = 0

array = [500, 100, 50, 10]

for coin in array :
  count = count + n // coin 
  n= n % coin

  print(count)
```

```python
n, k = map(int, input().split())

result = 0
while True : 
  target = (n//k) * k
  result = result + (n - target)
  n = target

  if n < k :
    break
  
  result = result + 1
  n = n//k

  result = result + (n-1)
  print(result)
```

```python
data = input()

result = int(data[0])

for i in range (1, len(data)):
  num = int(data[i])
  if num <= 1 or result <=1:
    result += num
  else:
    result *= num

print(result)
```

```python
def factorial_iterative(n):
  result = 1
  for i in range(1, n+1):
      result *= i
  return result

def factorial_recursive(n):
    if(n<=1):
      return 1
    return n * factorial_recursive(n-1)

print('반복적으로 구현:', factorial_iterative(5))
print('재귀적으로 구현', factorial_recursive(5))
```
